```python
import pandas as pd
import numpy as np

# Set number of characters
num_characters = 50  # Control the number of characters here

# Generate random values for the factors
np.random.seed(42)  # Set a seed for reproducibility
data = {
    #'id': [f'Char{i}' for i in range(num_characters)],
    'Species': np.random.choice(['Aquari', 'Zorblax', 'Florian', 'Faerix', 'Nexoon', 'Mythron','Emotivor', 'Sentire', 'Quixnar', 'Cybex'], num_characters),
    'Self-interest (S)': np.random.rand(num_characters),
    'Erosion of Trust (T)': np.random.rand(num_characters),
    'Perceived Injustice (I)': np.random.rand(num_characters),
    'Power Dynamics (P)': np.random.rand(num_characters),
    'Fear of Loss (F)': np.random.rand(num_characters),
    'Opportunity (O)': np.random.rand(num_characters),
    'Clan Intermingling (C)': np.random.rand(num_characters),
    'Greed (G)': np.random.rand(num_characters),
    'Temptation (Te)': np.random.rand(num_characters),
    'Poor Respect (R)': np.random.rand(num_characters),
    'Emotional Submission (Es)': np.random.rand(num_characters)
}

# Create DataFrame
df = pd.DataFrame(data)

# Function to calculate Probability of Betrayal
def calculate_betrayal_probability(row):
    I_squared = row['Perceived Injustice (I)'] ** 2
    P_squared = row['Power Dynamics (P)'] ** 2
    Te_squared = row['Temptation (Te)'] ** 2

    # Formula for Probability of Betrayal
    betrayal_prob = (
        row['Self-interest (S)'] +
        row['Erosion of Trust (T)'] +
        I_squared +
        P_squared +
        row['Fear of Loss (F)'] +
        row['Opportunity (O)'] +
        row['Clan Intermingling (C)'] +
        row['Greed (G)'] +
        Te_squared +
        row['Poor Respect (R)'] -
        row['Emotional Submission (Es)']
    ) / 10

    return betrayal_prob

# Apply the formula to calculate the probability for each character
df['Probability of Betrayal'] = df.apply(calculate_betrayal_probability, axis=1)

# Display the DataFrame
df.head()

# Optionally, save to CSV
df.to_csv('betrayal_probabilities.csv', index=False)
```
